#Aplikasi menghitung luas persegi panjang

```python
panjang = int(input("masukkan panjang dari persegi panjang (cm): "))
lebar = int(input(" masukkan lebar dari persegi panjang (cm) "))

luas_persegi_panjang =panjang * lebar

print(f"luas persegi panjang yaitu {luas_persegi_panjang} cm^2.")
```

#Aplikasi menghitung Volume Balok

```python
panjang = int(input("masukkan panjang dari balok (cm): "))
lebar = int(input(" masukkan lebar dari balok (cm) "))
tinggi = int(input(" masukan tinggi dari balok (cm): "))

volume_balok = panjang*lebar*tinggi

print(f"volume balok yaitu {volume_balok} cm^3.")
```
